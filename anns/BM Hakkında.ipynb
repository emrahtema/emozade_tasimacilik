# Boltzmann Machines

<br>
Boltzmann makinelerinin ağ görüntüsü resimdeki gibidir. Öğretileceği verilerdeki her bir parametreyi birbirleriyle ilişkili olabileceklerini varsayarak ilişkilendirir ve bu ilişkiler öğrendiği bilgilere dayanarak güçlenebilir yada zayıflayabilir. Her bir durumu değerlendirerek çalıştığı için en güçlü unsupervized modellerinden biridir. Resimde iki farklı nöron tipi görünse de makineye göre hepsi birbiriyle eşittir.

### Energy-Based Models
Boltzmann makineleri en düşük enerji durumu ilkesine dayanarak çalışırlar. Örnek vermek gerekirse insanların yaşadığı bir ev düşünelim ve evin içinde hava var. Hava moleküllerinin evin içine dağılımı normaldir, ancak bütün hava moleküllerinin tek bir odaya toplanabilme durumu mevcuttur. Bu durumda diğer odalardaki insanlar havasız kalıp ölürler. Ama neden böyle bir şey olmuyor, çünkü bir evi dolduracak hava molekülleri bir odaya toplanırsa birbirlerine çok yakın olacaklarından enerji artışı olacaktır, kaos tarzı bir şey yaşanacaktır. Çünkü hava moleküllerinin en normal durumu kafalarına göre dağınık oldukları durumdur. Boltzmann makineleri de bunu diyor, hava molekülleri tüm eve dağılıyor çünkü en az enerji harcadıkları sistem budur. Bir suya mürekkep damlası damlatmak gibi, damla olarak kalmaz suya dağılır çünkü en düşük enerji durumları budur, mürekkep molekülleri bir arada da kalabilir bu mümkündür ama kalmıyorlar. Bir suya zeytinyağı atsak suya karışmaz suyun üzerinde molekülleri bir arada durur çünkü zeytinyağının en düşük enerji durumu budur. Boltzmann makineleri de en düşük enerji durumunu tespit edecek bir dağılım formülüyle çalışırlar. Bu enerji de nöronlar arasındaki bağlantıların ağırlıkları olarak hesaplanır. Boltzmann makine ağı eğitildikten sonra da ortada bir sürü durum oluşacaktır ancak makine en uygun durumu bulmaya çalışacaktır.

### Restricted Boltzmann Machines
<br>
Boltzmann makinelerini implement etmek çok zordur çünkü nöron sayısı arttıkça aralarındaki ilişki sayısı da katlanarak artıyor. Restricted BM'de ise resimde görüleceği üzere bağlantı sayısını azaltmak için aynı renkli nodelar kendi aralarında ilişkisiz halde bulunuyorlar.

<br>

Yukarıdaki resimle RBM'nin çalışması açıklanabilir. Filmler visible nodelarda temsil ediliyor olsun, hidden nodelarda da filmlerin featureleri olsun. Bir kişinin beğendiği filmlere 1, beğenmediklerine 0, henüz izlemediklerine boş verilsin. Ağ yukarıdaki şekilde kişinin beğendiği filmlerin özelliklerine bakarak hidden nodeları düzenler. Kişinin beğendiği filmlerin featureleri olan nodelar yeşil yanarken, beğenmediği filmlerin featurelerine ait nodelar kırmızı yanıyor ve bu şekilde ağ güncelleniyor. Sonrasında bu kişinin Fight Club'u beğenip beğenmeme ihtimali hesaplansın. Yukarıdaki basit ağda FC'ye ait olan tek bir feature var o da Action türü olması, diğerlerini sağlamıyor. Ağ bütün hidden nodelara bakıyor ve bu nodeun filmle ilişkisi olduğunu tespit ediyor. İlişkisi ise kırmızı renkli ve bu yüzden BM, kişinin bu filmi beğenmeyeceğine karar veriyor. Bir de The Departed filmine baksın. O da Drama, Action, Dicaprio ve Oscar featurelerine sahip. Bunlara bakıyor ve 3'ünün yeşil, birinin kırmızı olduğunu tespit ediyor ve kişinin bu filmi beğeneceğine karar veriyor.

### Contrastive Divergence
Restricted BM'nin öğrenmesini sağlayan algoritmadır. Normal sinir ağlarında backpropagation yaparak öğrenme işlemini sağlayıp lossun en düşük hale gelmesine uğraşıyorduk. BM'ler direkt ağlar olmadıkları için ağırlıkların ANN'lerden farklı olarak güncellenmesi gerekiyor ki burada devreye bu algoritma CD giriyor.

### Deep Belief Networks
<br>
Birbiri ardına inşa edilmiş RBM'lerin oluşturduğu ağa denir. Mavi renkli giriş nodeları sabittir, ancak hidden kırmızı renkli nodelar birkaç farklı katman olabilirler. En alt katmandaki kırmızı katman mavilere doğru direkt olarak bağlantılara sahiptir, yukarıdakiler ise direkt değillerdir. Yukarıdaki resimde kırmızılı 3 katman vardır ki bu 3 tane RBM olduğunu gösteriyor. Deep Belief Networks'leri eğitmek için iki farklı algoritma vardır. Birisi "Greedy Layer-Wise", bu yukarıdaki 3 RBM'li ağı eğitmede kullanabileceğimiz bir algoritmadır. Eğitim katman katman gerçekleşir. Diğer algoritma "Wake-Sleep" algoritmasıdır.

### Deep Boltzmann Machines
Deep Belief Networks'lerden farklıdır, aynı şey değillerdir. Farkları hiçbir katmanın bir katmana direkt olmamasıdır, bütün katmanlardaki bütün nodelar undirected'dir.<br>
