# Auto Encoders

<br>
Auto Encoders resimde görüldüğü gibi featuresler giriş nodelarını oluşturur ve çıkış nodeları da giriş nodelarının aynısıdır. Bütün işlem soldan sağa doğru, ortadaki hidden layerden geçerek gerçekleşir. Outputlara sahip olduğu için bu Auto Encoders'lere yarı unsupervised denilir çünkü sınıflar-labeller olmadan eğitim görüyor ama bir çıktı oluşturuyor. Verilerden features çıkarımı için kullanılabilirler, öneri-tavsiye sistemleri için kullanılabilirler ve isimlerinden de anlaşılacağı üzere encoder-decoder olarak verileri daha küçük parçalara getirebilirler.

### Training an AutoEncoder
1- Bir array ile başla. Arrayde satırlar kullanıcıları ifade ediyor, sütunlarsa her bir filmi temsil ediyor. Her bir sütunun içerdiği bilgi o kullanıcının o filme verdiği skordur.<br>
2- İlk satır-kullanıcı networke girer, bütün filmlere verdiği skor bilgileriyle birlikte. Bu veri X vektörü olsun.<br>
3- X vektörü bazı mapping teknikleriyle Z vektörüne encode edilsin-dönüştürülsün ki Z vektörünün boyutu daha düşük olacaktır encode edildiği için.<br>
4- Z vektörü, X vektörüyle aynı derinliğe-dimensionsa sahip olan çıkış vektörü Y'ye decode edilir.<br>
5- Reconstraction error hesaplanır ki bu da loss fonksiyonuna karşılık gelir, bunun minimize edilmesi hedeflenir.<br>
6- Sağdan sola doğru backpropagation uygulanır.<br>
7- 1-6 adımlarını tekrarla. Her bir gözlemden-obzervationdan sonra ağırlıkları güncelle(Reinforcement Learning), yada ağırlıkları sadece her bir küme-batchın gözleminden sonra güncelle(Batch Learning).
8- Hepsi tamamlanınca epoch tamamlanmış oluyor. Tüm epochları gerçekleştir.

### Overcomplete Hidden Layers
Hidden layerdeki node sayısının giriş node sayısından daha fazla olmasını istersek ne olacak? Verilerden daha fazla feature çıkarılmasını hedeflemiş olabiliriz ama bu durumda her bir girişe karşılık gelecek şekilde saklı katmanda bir node belirlenecek ve her bir eşleşmedene sonra hidden sayısı daha fazla olduğu için boşta kalan hidden layer nodeleri olacaktır. Bu da ağın başarısız olması ve işe yarar bir bilgi çıkaramayacağımız anlamına gelir.

### Sparse AutoEncoders
<br>
Hidden layerdeki node sayısı girişlerden fazla olan encodersdir. Sparsity içeren regularization teknikleri kullanır. Bu teknik yukarıda bahsedilen sorunu çözer, o yüzden en sık kullanılan AutoEncoders algoritması-yapısı budur.

### Denoizing AutoEncoders
Bir başka regularization tekniğidir. Bu teknik visible giriş nodelarını alır ve onların yerine başka nodelar koyar. Yani input valuelerde değişiklik yapar. Çıkış nodeları ama orjinal veriler olarak kalırlar değişmezler.

### Contractive AutoEncoders
Bu da yukarıdaki sorunu çözmek için kullanılan bir regularization tekniğidir. Loss fonksiyonu üzerinde bir takım eklemeler yaparak backpropagation işlemini düzgünleştirir ve overfittingin engellenmesine yardımcı olur.

### Stacked AutoEncoders
<br>
Normal AutoEncoders ağına birden fazla hidden layer eklenmesiyle oluşur. Encoder etme kısımları artmış olur. Saklı katmanlarının sayısı artmasına rağmen bu normal bir AutoEncoder'dir, derin bir AE değildir.

### Deep AutoEncoders
<br>
Restricted Bolzman Machines'lerden türeme bir auto encoderdir. Zaten ağ şekline baktığımızda RBM'ye benzediğini hemen anlayabiliyoruz.
