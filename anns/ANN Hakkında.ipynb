# Artificial Neural Networks

### The Neuron

Yapay sinir ağlarındaki nöronların içinde neler gerçekleşir? İnput1,2,3..n -> Neuron -> Output<br>
Bütün girişlerin değerleri ağırlıklar ile çarpılıp toplanır.<br>
Sonra aktivasyon fonksiyonu uygulanır.<br>
Fonksiyon uygulandıktan sonra nöron bir değer çıkışı verip vermeyeceğine karar verir.
(Nöron çıktısı kategorik, binary yada değişken bir sayı olabilir)

### The Activation Function

Bazı aktivasyon fonksiyonları şunlardır:<br>
<br>
Bu fonksiyonun uygulandığı nöron içindeki bilgi sıfırdan küçükse çıktı sıfır, büyükse 1 olur. Yani binary bir çıkış gerçekleşir.

<br>
Bu fonksiyon genelde linear regresyon tarzı modellerde kullanılır. Çıktısı değişkendir ve sıfırın altı değerleri geçirmez, sıfırın üstü ne varsa onu verir. (X = nöron içindeki değer, diğer fonksiyonlarda da aynı şekilde)

<br>
En çok kullanılan fonksiyonlardan biridir. Fonksiyon sıfırda başlar ve nöron içindeki değere göre yükselip çıktı verir.

<br>
Sigmoid fonksiyonunun benzeridir, farkı -1 ile 1 arasında bir çıktı verir.

ÖRNEK: Eğer nöron çıkışlarımız binary olacaksa Treshhold fonksiyonu tercih edilmelidir. Tabi duruma göre Sigmoid de tercih edilebilir çünkü o da sıfırın altı değerleri sıfır kabul edeceğinden mümkün olabilir.<br>
Saklı katmanlarda ve output katmanında farklı aktivasyon fonksiyonları kullanılabilir. Örneğin bir tane saklı katmandan oluşan ağımız var, saklı katmanda Rectifier kullanılırken, çıkış katmanında Sigmoid kullanılabilir.

### Yapay Sinir Ağları Nasıl Çalışır?

Saklı katmanlardaki her bir nöron giriş değerlerine göre şekillenir. Hangi giriş değerlerinin birbiriyle ilişkili olduğuna yada sonuca etkisine göre bir karara varıp ona uygun davranır. Örneğin 4 tane nöronu olan bir gizli katmanımız var ve ev fiyatı tahmini için eğitiyoruz. İlk nöron metrekaresi, odasayısı ve merkezeuzaklığı girişlerinin önemli olduğunu görüyor, yaş bilgisinin önemsiz olduğunu görüyor bu üçü varken ve buna göre sadece bu üçünü hesaplayacak şekilde ayarlanıyor. İkinci nöron da merkeze uzaklığı önemli olmayan veri girişleri tespit ediyor ve buna göre diğer 3 özellikle şekilleniyor. Üçüncü nörün evin yaşının kendi başına fiyatı belirleyen bir etken olduğunu tespit ediyor, yaş büyüdükçe fiyat düşüyor diye ayarlanıyor. Son nöron da bakıyor ki bazı verilerde evin yaşı 100'den büyük olunca fiyatı artmış diyorki 100 yaşından büyük evlerin fiyatı yükseliyor ona göre kendini ayarlıyor, 100 yaşın altındaki evlerle ilgilenmiyor bile. Buna göre gizli katmandaki her bir nöron şekilleniyor ve çalışıyor. O yüzden nöronlar tüm değerlerden sadece birini de kabul edebilir, hepsini de yada birkaçını da. Bu yüzden bir nöron ev fiyatını kendi başına doğru tahmin edemez ama bu ayarlamaları sağlamış nöronlar topluluğu evin fiyatını başarıyla tahmin edebilir. Gizli katmanlardaki nöron sayısını değiştirmek bu yüzden farklı sonuçlar verir. 

### ANN Nasıl Öğrenir? - Cost Function

Bir yapay sinir ağı girişlere göre şekillenen saklı katmanlar arasında ağırlıklar tespit eder ve her bir nöron bunlara göre şekillenip yapay sinir ağını oluşturur. Oluşum bittikten sonra Cost function hesaplanır, yani verilerin kaç tanesinin sınıfı doğru tahmin edilebiliyor.<br>
y = asıl sınıf<br>
y^ = tahmin edilen sınıf<br>
C = 1/2(y^-y)^2 (bu cost fonksiyonlarından birisi sadece)<br>
Eğer cost fonksiyonu büyükse bu bilgi ağa geri gider ve ağırlıklar bu sonuca göre güncellenir C değerini düşürebilmek için. Bu değer ne kadar düşerse, yapay sinir ağı o kadar doğru çalışır. En düşük değeri bulunduğunda bu bizim tamamlanmış sinir ağımızdır. Bu cost fonksiyonuna göre ağırlıkların yeniden belirlenmesi işlemine de Back Propagation deniliyor.

### Gradient Descent

<br>
Gradient Descent işlemi, yukarıdaki resimdeki gibi bir çok tekrar ve deneme sonucu Cost fonksiyonun değerinin en aşağıya çekilebilmesidir, yani en iyi değer, en düşük olanı elde etmektir.<br>

<br>
Bir nokta alınıp o noktadan C hesaplanır ve çıkan sonuca göre zikzaklar çizilerek bölge sürekli taranır ve ortaya en yakın konuma gelinince durulur. Resimde zikzaklar olarak görünüyor ama aslında tarama işlemi yapılıyor sanki top oval bir yüzeyde aşağı yukarı gidiyormuş gibi gücü bitince en ortada duruyormuş gibi hayal edilebilir, bu bizim gradient descent işlemimizin hareketini gösterir. Cost fonksiyonu seçerken gradient descentin hareketine uygun olarak seçmek gerekir, yani konveks olmalı. Konvex değil, yani yukarıdaki gibi U şeklinde değil de dalgalı bir yapı varsa ne yapılmalı peki?

<br>
Konveks değil de yukarıdaki gibiyse en iyi yeri değil, önüne gelen en uygun yeri seçecektir, bu yüzden bu konveks yapılmalı çünkü yukarıdakinde en doğru ağırlıklar hesaplanamayacaktır, yani optimize edilmiş bir yapay sinir ağımız olmayacaktır.<br>
Burada kullanılması gereken şey Stochastic Gradient Descent olacaktır.<br>
<b>Stochsatic Gradient Descent</b>, Cost fonksiyonunun konvex olmasını gerektirmiyor. SGD, normal GD (Batch Gradient Descent yukarıdaki)'den biraz farklı çalışır. Gradient Descent'de ağ kurulduktan sonra bütün veriler birden test edilir ve Cost fonksiyonun sonucuna göre ağırlıklar yeniden hesaplanırdı. SGD'de ise bütün veriler birden test edilmiyor, her bir veri satır satır-tek tek test ediliyor ve her birinin Cost fonksiyonu sonucuna göre ağırlıklar güncelleniyor. Daha uzun vakit gerektiriyor gibi görünse de aslında daha hızlıdır çünkü tüm veriyi memoriye yüklemekle ve hepsinin çalışmasını beklemekle uğraşmıyor.

<br>
Aynı giriş verileri olduğunda Gradient Descent metodu her iterasyonda aynı işlemleri yapar ve aynı sonuçlara ulaşır, aynı sinir ağını bulur. SGD ise her bir veriyi random olarak seçip Cost hesaplayıp ağırlık belirlediğinden, yapay sinir ağı her kurulduğunda farklı bir ağ kurulmuş oluyor. Bu da yerine göre bir dez avantaj olarak görülebilir. Bu iki yapının arasında kalan bir yapı daha vardır ki buna da Mini Batch Gradient Descent denilir. Bu da ikisinin benzeri gibi çalışır, ne bütün veriyi birden işler, ne de tek tek işler. Veriyi kümeler halinde alıp işler, Cost fonksiyonunu hesaplar ve ona göre ağırlıkları günceller. Örneğin bu veri kümeleri 5, 10, 50, 100, 1000'li kümeler olabilir.

### Back Propagation

Gradient Descent'de Cost fonksiyonu sonucuna göre ağırlıkları güncelleyip yapay sinir ağını yeniliyorduk. Back Propagation'un işlevi Cost'un büyümesine sebep olan hatalara hangi ağırlıklar sebep oluyor onları tespit etmek, böylece bütün ağırlıkları güncellerken, hangi ağırlık hataya ne kadar sebep oluyorsa bunu tespit edebildiğinden, hataya sebep olduğu kadar bir güncelleme sağlanabiliyor. Hataya çok fazla sebep olan bir ağırlık daha büyük bir miktarda güncellenirken, hataya az sebep olan bir ağırlığın güncellenen değeri önceki değerinden azıcık farklı olacaktır.<br><br>
