### ANN Nasl renir? - Cost Function

Bir yapay sinir a girilere gre ekillenen sakl katmanlar arasnda arlklar tespit eder ve her bir nron bunlara gre ekillenip yapay sinir an oluturur. Oluum bittikten sonra Cost function hesaplanr, yani verilerin ka tanesinin snf doru tahmin edilebiliyor.<br>
y = asl snf<br>
y^ = tahmin edilen snf<br>
C = 1/2(y^-y)^2 (bu cost fonksiyonlarndan birisi sadece)<br>
Eer cost fonksiyonu bykse bu bilgi aa geri gider ve arlklar bu sonuca gre gncellenir C deerini drebilmek iin. Bu deer ne kadar derse, yapay sinir a o kadar doru alr. En dk deeri bulunduunda bu bizim tamamlanm sinir amzdr. Bu cost fonksiyonuna gre arlklarn yeniden belirlenmesi ilemine de Back Propagation deniliyor.

### Gradient Descent

<br>
Gradient Descent ilemi, yukardaki resimdeki gibi bir ok tekrar ve deneme sonucu Cost fonksiyonun deerinin en aaya ekilebilmesidir, yani en iyi deer, en dk olan elde etmektir.<br>

<br>
Bir nokta alnp o noktadan C hesaplanr ve kan sonuca gre zikzaklar izilerek blge srekli taranr ve ortaya en yakn konuma gelinince durulur. Resimde zikzaklar olarak grnyor ama aslnda tarama ilemi yaplyor sanki top oval bir yzeyde aa yukar gidiyormu gibi gc bitince en ortada duruyormu gibi hayal edilebilir, bu bizim gradient descent ilemimizin hareketini gsterir. Cost fonksiyonu seerken gradient descentin hareketine uygun olarak semek gerekir, yani konveks olmal. Konvex deil, yani yukardaki gibi U eklinde deil de dalgal bir yap varsa ne yaplmal peki?

<br>
Konveks deil de yukardaki gibiyse en iyi yeri deil, nne gelen en uygun yeri seecektir, bu yzden bu konveks yaplmal nk yukardakinde en doru arlklar hesaplanamayacaktr, yani optimize edilmi bir yapay sinir amz olmayacaktr.<br>
Burada kullanlmas gereken ey Stochastic Gradient Descent olacaktr.<br>
<b>Stochsatic Gradient Descent</b>, Cost fonksiyonunun konvex olmasn gerektirmiyor. SGD, normal GD (Batch Gradient Descent yukardaki)'den biraz farkl alr. Gradient Descent'de a kurulduktan sonra btn veriler birden test edilir ve Cost fonksiyonun sonucuna gre arlklar yeniden hesaplanrd. SGD'de ise btn veriler birden test edilmiyor, her bir veri satr satr-tek tek test ediliyor ve her birinin Cost fonksiyonu sonucuna gre arlklar gncelleniyor. Daha uzun vakit gerektiriyor gibi grnse de aslnda daha hzldr nk tm veriyi memoriye yklemekle ve hepsinin almasn beklemekle uramyor.

<br>
Ayn giri verileri olduunda Gradient Descent metodu her iterasyonda ayn ilemleri yapar ve ayn sonulara ular, ayn sinir an bulur. SGD ise her bir veriyi random olarak seip Cost hesaplayp arlk belirlediinden, yapay sinir a her kurulduunda farkl bir a kurulmu oluyor. Bu da yerine gre bir dez avantaj olarak grlebilir. Bu iki yapnn arasnda kalan bir yap daha vardr ki buna da Mini Batch Gradient Descent denilir. Bu da ikisinin benzeri gibi alr, ne btn veriyi birden iler, ne de tek tek iler. Veriyi kmeler halinde alp iler, Cost fonksiyonunu hesaplar ve ona gre arlklar gnceller. rnein bu veri kmeleri 5, 10, 50, 100, 1000'li kmeler olabilir.

### Back Propagation

Gradient Descent'de Cost fonksiyonu sonucuna gre arlklar gncelleyip yapay sinir an yeniliyorduk. Back Propagation'un ilevi Cost'un bymesine sebep olan hatalara hangi arlklar sebep oluyor onlar tespit etmek, bylece btn arlklar gncellerken, hangi arlk hataya ne kadar sebep oluyorsa bunu tespit edebildiinden, hataya sebep olduu kadar bir gncelleme salanabiliyor. Hataya ok fazla sebep olan bir arlk daha byk bir miktarda gncellenirken, hataya az sebep olan bir arln gncellenen deeri nceki deerinden azck farkl olacaktr.<br><br>
