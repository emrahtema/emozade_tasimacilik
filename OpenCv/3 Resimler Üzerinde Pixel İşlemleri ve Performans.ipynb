# 1 Pixellere Eri≈ümek ve Oynamak

```python
img = cv.imread('sabertooth.jpg')
px = img[100,100]
print(px)

# Sadece mavi pixele eri≈ümek. Blue 0, Red 2, Green 1 
blue = img[100,100,0]
print(blue)
img.itemset((100,100,2),100)# kƒ±rmƒ±zƒ± renk deƒüerini deƒüi≈ütirelim
print(img.item(100,100,2))

# Pixel deƒüerlerini de aynƒ± ≈üekilde deƒüi≈ütirebiliyoruz
img[100,100] = [255,255,255]

print("----")
# Resmin √∂zelliklerine eri≈üebiliriz
print(img.shape)
print(img.size)
print(img.dtype)

print("----")
# Resimdeki bir takƒ±m pixelleri ba≈üka yere ta≈üƒ±yabiliriz
secilen_alan = img[300:500, 300:400]
img[100:300, 100:200] = secilen_alan
#plt.subplot(121); plt.imshow(img)

# Mavi kanallarƒ± sƒ±fƒ±r yapalƒ±m
img[:,:,0] = 0
plt.subplot(121); plt.imshow(img)

#------------------------------
# 2'de Resim birle≈ütirme, birbirine ekleme, belirli bir rengi silip saydam yapma gibi √∂zellikler var.
# 3'te OpenCV i≈ülemlerinde performansla ilgili ≈üeyler var.
```
