```python
import h5py
from PIL import Image

fileName = 'data.h5'
numOfSamples = 17568
with h5py.File(fileName, "w") as out:
    out.create_dataset("X", (numOfSamples, 256, 256, 3), dtype='u1')
    out.create_dataset("Y", (numOfSamples, 1, 1), dtype='u1')       
```

```python
from glob import glob
import cv2

imagePatches = glob('chest_xray/**/**/*.jpeg', recursive=True)
print(len(imagePatches))

#pattern_normal = '*NORMAL*'
#pattern_bacteria = '*_bacteria_*'
#pattern_virus = '*_virus_*'

#normal = fnmatch.filter(imagePatches, pattern_normal)
#bacteria = fnmatch.filter(imagePatches, pattern_bacteria)
#virus = fnmatch.filter(imagePatches, pattern_virus)
x = []
y = []
for img in imagePatches:
    full_size_image = cv2.imread(img)
    im = cv2.resize(full_size_image, (256, 256), interpolation=cv2.INTER_CUBIC)
    x.append(im)
    if "NORMAL" in img:
        y.append(0)
    elif "bacteria" in img:
        y.append(1)
    elif "virus" in img:
        y.append(1)
    else:
        #break
        print('no class')
```

```python
import numpy as np

for i in range(len(x)):
    with h5py.File(fileName, "a") as out:
        out['X'+str(i)] = np.asarray(x[i])
        out['Y'+str(i)] = y[i]
```
